## Calculate the traffic with the GLCM texture features
- Energy is inversely proportional to traffic
- Entropy is directly proportional to traffic

```python
import cv2
import numpy as np
import math
import matplotlib.pyplot as plt
import utils
```

```python
# Read image as greyscale
img = cv2.imread('Traffic.jpg', cv2.IMREAD_GRAYSCALE)

fig, (ax1, ax2) = plt.subplots(figsize=(13,3), ncols=2)
ax1.imshow(img, cmap="gray")

energy, entropy, glcm = compute_GLCM(img)
pos = ax2.imshow(glcm,  cmap='hot')
cbar = fig.colorbar(pos, ax=ax2)
print("Energy: %.2f\tEntropy: %.2f" % (energy, entropy))
```
